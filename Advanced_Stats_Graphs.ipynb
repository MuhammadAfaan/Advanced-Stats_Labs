```python
import numpy as np  #For numerical operations
import pandas as pd  #For handling datasets
import matplotlib.pyplot as plt  #For data visualization
import seaborn as sns  #For advanced visualizations
from scipy import stats  #For statistical computations
```

```python
#Generate synthetic data
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=100)  #Mean=70, StdDev=10, n=100
data_df = pd.DataFrame(data, columns=["Scores"])
data_df.to_csv("scores.csv", index=False)
```

```python
#Read the dataset
data = pd.read_csv("scores.csv")

#Calculate descriptive statistics
mean = data["Scores"].mean()
median = data["Scores"].median()
mode = data["Scores"].mode()[0]
std_dev = data["Scores"].std()
variance = data["Scores"].var()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
```

```python
plt.hist(data["Scores"], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(data["Scores"], vert=False, patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.title("Boxplot of Scores")
plt.xlabel("Scores")
plt.show()
```

# **Task#2**

```python
# Load a dataset (Example: Student Performance dataset)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
#displaying rows
print(df.head())
```

```python
#Calculate descriptive statistics
mean = df["sepal_length"].mean()
median = df["sepal_length"].median()
mode = df["sepal_length"].mode()[0]
std_dev = df["sepal_length"].std()
variance = df["sepal_length"].var()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
```

```python
# get numerical coloum
print(df.describe())
```

```python
# Histogram
sns.histplot(df['sepal_length'], bins=20, kde=True)
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# Boxplot
sns.boxplot(x=df['sepal_length'])
plt.title('Boxplot of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Value')
plt.show()
```

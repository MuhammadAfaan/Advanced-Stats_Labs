```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sums = np.arange(2, 13)
probabilities = np.array([1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1]) / 36

cumulative_probabilities = np.cumsum(probabilities)
```

```python
#plotting distribution
plt.figure(figsize=(10, 5))
plt.bar(sums, probabilities, color='green', alpha=0.7, label='P(X)')
plt.xlabel("Sum of Two Dice")
plt.ylabel("Probability")
plt.title("Discrete Probability Distribution of Sum of Two Dice")
plt.xticks(sums)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# cumalitive plot
plt.figure(figsize=(10, 5))
plt.plot(sums, cumulative_probabilities, marker='o', linestyle='-', color='red', label='F(X)')
plt.xlabel("Sum of Two Dice")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Probability Distribution of Sum of Two Dice")
plt.xticks(sums)
plt.ylim(0, 1.1)
plt.legend()
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

# Task#**2**

```python
import scipy.stats as stats
from scipy.special import comb
```

```python
# parameters
total_balls = 10
black_balls = 6
red_balls = 4
sample_size = 5

# x number of black balls
x_values = np.arange(0, 6)

# hypergeometric formula
def hypergeometric_pmf(k, N, K, n):
    return (comb(K, k) * comb(N-K, n-k)) / comb(N, n)

probabilities = [hypergeometric_pmf(k, total_balls, black_balls, sample_size) for k in x_values]

# cum probabilites
cumulative_probabilities = np.cumsum(probabilities)
```

```python
# Plot Discrete Probability Distribution
plt.figure(figsize=(10, 5))
plt.bar(x_values, probabilities, color='blue', alpha=0.7, label='P(X)')
plt.xlabel("Number of Black Balls in Sample")
plt.ylabel("Probability")
plt.title("Discrete Probability Distribution (Without Replacement)")
plt.xticks(x_values)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot Cumulative Probability Distribution
plt.figure(figsize=(10, 5))
plt.plot(x_values, cumulative_probabilities, marker='o', linestyle='-', color='red', label='F(X)')
plt.xlabel("Number of Black Balls in Sample")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Probability Distribution (Without Replacement)")
plt.xticks(x_values)
plt.ylim(0, 1.1)
plt.legend()
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
```

```python
from scipy.stats import binom
```

```python
#parameters
num_questions = 10
prob_correct = 1/4
marks_per_question = 5

# expected correct
expected_correct = num_questions * prob_correct

# expected marks
expected_marks = marks_per_question * expected_correct

# varience
variance_correct = num_questions * prob_correct * (1 - prob_correct)

# varience marks
variance_marks = (marks_per_question ** 2) * variance_correct

print(f"Expected Correct Answers: {expected_correct}")
print(f"Expected Marks: {expected_marks}")
print(f"Variance of Marks: {variance_marks}")
```

# Task#**4**

```python
#parameters
num_questions = 10
probabilities = [1/4, 2/4, 3/4]
labels = ["1 correct option", "2 correct options", "3 correct options"]
colors = ['blue', 'green', 'red']

#skewness
skewness_values = [binom.stats(num_questions, p, moments='s') for p in probabilities]

#ploting distribution
x_values = np.arange(0, num_questions + 1)
plt.figure(figsize=(12, 5))

for i, p in enumerate(probabilities):
    pmf_values = binom.pmf(x_values, num_questions, p)
    plt.plot(x_values, pmf_values, marker='o', linestyle='-', color=colors[i],
             label=f"{labels[i]} (Skewness={skewness_values[i]:.2f})")

plt.xlabel("Number of Correct Answers")
plt.ylabel("Probability")
plt.title("Probability Distributions for Different Quiz Cases")
plt.legend()
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()

#printing results
for i, skew in enumerate(skewness_values):
    print(f"Case {i+1}: {labels[i]}")
    print(f"  Skewness: {skew:.2f}")
    if skew > 0:
        print("  Interpretation: Right-skewed (positively skewed)")
    elif skew < 0:
        print("  Interpretation: Left-skewed (negatively skewed)")
    else:
        print("  Interpretation: Symmetric (approximately normal)")
    print("-")
```

# Task#**5**

```python
from scipy.stats import binom, poisson, geom, nbinom, hypergeom
```

```python
# Part-1

n_students = 15
p_pass = 0.7

prob_at_least_10 = 1 - binom.cdf(9, n_students, p_pass)
print(f"Probability that at least 10 students pass: {prob_at_least_10:.4f}")

# plotting class sizes of 10,30,50,500

class_sizes = [10, 30, 50, 500]
plt.figure(figsize=(12, 6))
for n in class_sizes:
    x = np.arange(n + 1)
    pmf = binom.pmf(x, n, p_pass)
    plt.plot(x, pmf, marker='o', linestyle='-', label=f'n={n}')
plt.xlabel("Number of Students Passing")
plt.ylabel("Probability")
plt.title("PMF for Different Class Sizes (p=0.7)")
plt.legend()
plt.grid()
plt.show()
```

# Part-**2**

```python
lambda_visits = 5

# plotting poission dirtribution
x = np.arange(0, 15)
poisson_pmf = poisson.pmf(x, lambda_visits)
plt.figure(figsize=(10, 5))
plt.bar(x, poisson_pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Visitors")
plt.ylabel("Probability")
plt.title("Poisson Distribution of Website Visits (=5)")
plt.show()

# probability of 3 visitors
prob_3_visitors = poisson.pmf(3, lambda_visits)
print(f"Probability of exactly 3 visitors: {prob_3_visitors:.4f}")

# pmf for different user per minutes
lambda_values = [1, 5, 50]
plt.figure(figsize=(12, 6))
for lam in lambda_values:
    x = np.arange(0, lam * 3)
    pmf = poisson.pmf(x, lam)
    plt.plot(x, pmf, marker='o', linestyle='-', label=f'={lam}')
plt.xlabel("Number of Visitors")
plt.ylabel("Probability")
plt.title("Poisson PMF for Different Averages")
plt.legend()
plt.grid()
plt.show()
```

# Part-**3**

```python
# fraud detection with geometric mean
p_fraud = 0.1

# probability of first fraud at 6th attempt
prob_6th_attempt = geom.pmf(6, p_fraud)
print(f"Probability of detecting first fraud on 6th attempt: {prob_6th_attempt:.4f}")

# number of attempts to detect fraud
expected_attempts = 1 / p_fraud
print(f"Expected number of attempts to detect fraud: {expected_attempts:.2f}")

# PMF for diff fraud prob
fraud_probs = [0.05, 0.1, 0.2]
plt.figure(figsize=(12, 6))
x = np.arange(1, 20)
for p in fraud_probs:
    pmf = geom.pmf(x, p)
    plt.plot(x, pmf, marker='o', linestyle='-', label=f'p={p}')
plt.xlabel("Number of Attempts")
plt.ylabel("Probability")
plt.title("Geometric PMF for Different Fraud Probabilities")
plt.legend()
plt.grid()
plt.show()
```

```python
p_click = 0.2
r_clicks = 5

# number of users needed for 5 clicks
expected_users = r_clicks / p_click
print(f"Expected number of users needed for 5 clicks: {expected_users:.2f}")

# visualization with different probabilites
click_probs = [0.1, 0.2, 0.3]
x = np.arange(1, 50)
plt.figure(figsize=(12, 6))
for p in click_probs:
    pmf = nbinom.pmf(x, r_clicks, p)
    plt.plot(x, pmf, marker='o', linestyle='-', label=f'p={p}')
plt.xlabel("Number of Users")
plt.ylabel("Probability")
plt.title("Negative Binomial PMF for Different Click Probabilities")
plt.legend()
plt.grid()
plt.show()
```

```python
# hypergeometric distribution
N_laptops = 20
K_defective = 5
n_selected = 4

# prob at most 10 laptops are defective
print("At most 10 defective probability is 1 since max selection is 4.")

# expected n umber of defective laptops
expected_defective = n_selected * (K_defective / N_laptops)
print(f"Expected number of defective laptops: {expected_defective:.2f}")

# probability that 2 are defective
prob_2_defective = hypergeom.pmf(2, N_laptops, K_defective, n_selected)
print(f"Probability of selecting exactly 2 defective laptops: {prob_2_defective:.4f}")
```

# Part#**1**

```python
# Exponential Distribution (Bank Customers)
lambda_customers = 3 #3 customer per minute

# 1000 customers examples
exp_samples = np.random.exponential(1/lambda_customers, 1000)

# PDF plotting
x = np.linspace(0, 3, 1000)
pdf = lambda_customers * np.exp(-lambda_customers * x)
plt.figure(figsize=(10, 5))
plt.hist(exp_samples, bins=50, density=True, alpha=0.6, color='b', label='Simulated Data')
plt.plot(x, pdf, 'r-', label='Theoretical PDF')
plt.xlabel("Time between Arrivals (minutes)")
plt.ylabel("Density")
plt.title("Exponential Distribution of Customer Arrivals")
plt.legend()
plt.show()

# c) probability of waiting more than 30 seconds
prob_wait_30s = np.exp(-lambda_customers * 0.5)
print(f"Probability of waiting more than 30 seconds: {prob_wait_30s:.4f}")
```

```python
#normal distribution
mean_salary = 50000
std_salary = 10000

# generating 10k records
salary_samples = np.random.normal(mean_salary, std_salary, 10000)

# plotting pdf records
x = np.linspace(20000, 80000, 1000)
pdf = stats.norm.pdf(x, mean_salary, std_salary)
plt.figure(figsize=(10, 5))
plt.hist(salary_samples, bins=50, density=True, alpha=0.6, color='g', label='Simulated Data')
plt.plot(x, pdf, 'r-', label='Theoretical PDF')
plt.xlabel("Salary ($)")
plt.ylabel("Density")
plt.title("Normal Distribution of Employee Salaries")
plt.legend()
plt.show()

# probability of earning more than 60k
prob_above_60k = 1 - stats.norm.cdf(60000, mean_salary, std_salary)
print(f"Probability of earning more than $60,000: {prob_above_60k:.4f}")
```
